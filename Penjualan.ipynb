```python
import pandas as pd

df = pd.read_csv("dataset_penjualan.csv")

df
```

# Bagian Baru

```python
#Berapa total penjualan (dalam Rupiah) pada dataset ini?

total_penjualan = (df['Quantity'] * df['Price_per_unit']).sum()

print("Rp.", total_penjualan)
```

```python
# Kategori produk mana yang memiliki total penjualan tertinggi?
kategori_penjualan_tertinggi = df.groupby('Category')['Price_per_unit'].sum()

print("Kategori penjualan tertinggi adalah : ", kategori_penjualan_tertinggi)
```

```python
# Hitung rata-rata umur pelanggan untuk setiap kategori produk.
rata_umur_pelanggan = df.groupby('Category')['Customer_Age'].mean()

print('Rata-rata umur pelanggan : ', rata_umur_pelanggan)
```

```python
# Berapa rata-rata jumlah produk yang dibeli per transaksi?

rata_produk_pertransaksi = df.groupby('Product')['Quantity'].mean()

print('Rata-rata jumlah produk yang dibeli per transaksi:', rata_produk_pertransaksi)
```

```python
#Berapa total penjualan untuk pelanggan yang berusia di atas 30 tahun

df_30 = df[df['Customer_Age'] > 30]
total_penjualan = sum(df_30['Quantity'] * df_30['Price_per_unit'])
total_penjualan
```

```python
# Berapa jumlah pesanan dari setiap negara?

pesanan_tiap_negara = df.groupby('Customer_Country')['Quantity'].sum().sort_values(ascending=False)

print('jumlah pesanan dari setiap negara', pesanan_tiap_negara)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_group = df.groupby('Date')['Quantity'].sum().reset_index()

# Buatlah visualisasi (grafik) yang menunjukkan tren jumlah pesanan berdasarkan tanggal.
sns.lineplot(data = df_group, x='Date' , y='Quantity')
plt.title("tren jumlah pesanan berdasarkan tanggal")
plt.xticks(rotation = 30)
plt.show()
```
