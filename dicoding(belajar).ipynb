```python
import pandas as pd

df_order = pd.read_csv("orders.csv")

order_kolom = ['order_date', 'delivery_date']

for kolom in order_kolom :
  df_order[kolom] = pd.to_datetime(df_order[kolom])

df_order.info()
```

```python
delivery_time = df_order["delivery_date"] - df_order["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
df_order["delivery_time"] = round(delivery_time/86400)

df_order.describe(include = "all")
df_order
```

```python
customer_id_in_order_df = df_order.customer_id.to_list()
df_customer["status"] = df_customer["customer_id"].apply(lambda x : "Active" if x in customer_id_in_order_df else "Non active")
df_customer.sample(5)
```

```python
df_customer.groupby(by = "status").customer_id.count()
```

```python
order_customer_df = pd.merge(
    left = df_order,
    right = df_customer,
    how = "left",
    left_on="customer_id",
    right_on="customer_id"
)

order_customer_df.head()
```

```python
order_customer_df.groupby(by = "city").order_id.nunique().sort_values(ascending = False).reset_index().head(10)
```

```python
order_customer_df.groupby(by = "state").order_id.nunique().sort_values(ascending = False)
```

```python
order_customer_df.groupby(by = "gender").order_id.nunique().sort_values(ascending = False)
```

```python
order_customer_df["age_group"] = order_customer_df.age.apply( lambda x : "youth" if x <= 24 else ("senior" if x > 64 else "adults") )
order_customer_df.groupby(by = "age_group").order_id.nunique().sort_values(ascending = False)
```
