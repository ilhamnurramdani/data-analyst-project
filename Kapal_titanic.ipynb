```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder

df = sns.load_dataset('titanic')
df
```

```python
kategori = ['muda', 'remaja', 'tua']

df['kategori'] = pd.qcut(df['age'], q=3, labels=kategori)

df
```

```python
bins = np.linspace(min(df['age']), max(df['age']), 4)
df['kategori2'] = pd.cut(df['age'], bins=bins, labels = kategori)
df
```

```python
df['age'].fillna(df['age'].median(), inplace = True)
df.isnull().sum()
```

```python
#Siapa yang lebih banyak selamat? Laki-laki atau perempuan?

sns.countplot(data = df, x = 'survived', hue = 'sex')
plt.title('Penumpang selamat berdasarkan gender')
plt.show()
```

```python
#Apakah orang kaya (Pclass 1) punya kemungkinan lebih besar selamat?
sns.countplot(data = df, x = 'survived', hue = 'pclass')
plt.title('penumpang selamat berdasarkan class')
plt.show()
```

```python
#Apakah umur memengaruhi kemungkinan selamat?

sns.boxplot(data = df , x = 'survived', y = 'age')
plt.show()
```

```python
#Apakah orang yang bawa keluarga lebih besar peluangnya selamat?

sns.countplot(data = df, x = 'survived', hue = 'alone')
plt.title('Orang yang membawa keluarga')
plt.show()
```

```python
#Penumpang dari embarkasi mana paling banyak selamat?

sns.countplot(data = df, x = 'survived', hue = 'embarked')
plt.title('Penumpang dari embarkasi mana paling banyak selamat')
plt.show()
```
