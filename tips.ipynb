```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df
```

```python
#Tampilkan semua transaksi hari Minggu dengan tip lebih dari $5
df[(df['day'] == "Sun") & (df['tip'] > 5)]
```

```python
#Hitung rata-rata total bill pelanggan perempuan
df[df['sex'] == 'Female']['total_bill'].mean()
```

```python
#Hari apa yang punya total tip paling banyak?
df.groupby('day')['tip'].sum().sort_values(ascending=False)
```

```python
#Siapa yang lebih sering kasih tip besar: smoker atau non-smoker?
df.groupby('smoker')['tip'].sum()
```

```python
#Bikin histogram tip dengan 15 bin
sns.histplot(data = df['tip'], bins = 15 )
plt.title('Distribusi tip')
plt.show()
```

```python
#Bikin barplot rata-rata total_bill per gender
sns.barplot(x=df['sex'], y = df['total_bill'])
plt.title('Boxplot tip berdasarkan gender')
plt.show()
```

```python
#Scatterplot tip vs total_bill khusus pelanggan hari Minggu
sns.scatterplot(x='tip', y='total_bill', hue='sex', data=df[df['day'] == "Sun"])
plt.title('Scatterplot tip vs total_bill khusus pelanggan hari Minggu')
plt.show()
```
